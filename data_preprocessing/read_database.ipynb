```python
"""Extracting and summing all labels into a record dict"""

# Setting up the root directory
#rootdir = './database/european-st-t-database-1.0.0'
#rootdir = './database/mit-bih-arrhythmia-database-1.0.0'
rootdir = './database/mit-bih-st-change-database-1.0.0'     
#rootdir = './database/sudden-cardiac-death-holter-database-1.0.0'

files = os.listdir(rootdir)
name_list = []
#TODO: Choose the type of lead data want to be read
ECG = []
type = {}
sig_names = {}
#print(files)
for file in files:
    #TODO: Change the length of the data file name
    if file[0:3] in name_list:      # change the length to files name length
        continue
    else:       # exclude any repeated files
        name_list.append(file[0:3])     # same as above
#print(name_list)

#TODO:Carefully examine each database for specific conditions!!!

for name in name_list:
    # look at the second letter whether a numerical or not for ecg data
    if name[1] not in ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']:   # discard any non ecg files in the folder
        continue
    record = wfdb.rdrecord(rootdir + '/' + name)
    print(record.sig_name)               # Look for the highest frequency name for the record
    # counts for what methods used the most
    # for i in record.sig_name:
    #     if record.sig_name[i] not in sig_names.keys:
    #         sig_names[record.sig_name[i]] = 1
    #     else:
    #         sig_names[record.sig_name[i]] += 1
    
    #TODO: Change the most often lead
    if 'ECG' in record.sig_name:       # Pick one recording type (the most often)
        ECG.append(name)             
    annotation = wfdb.rdann(rootdir+'/'+name, 'atr')  # read .art
    for symbol in annotation.symbol:            # read all label types
        if symbol in list(type.keys()):
            type[symbol]+=1
        else:
            type[symbol]=1
    print('sympbol_name',type)

#TODO: CHANGE!
print(ECG) # test for which being included
sorted(type.items(),key=lambda d:d[1],reverse=True)     # Suming all types of label in the dataset
```

```python
'''Segmentation and adding labels to each segments and store in pd.DataFrame'''

#TODO: must change sampling frequency here
f=360       # sampling frequency
segmented_len=10        # cut into 10s segments
label_count=0
count=0
abnormal=0

segmented_data = []             # X
segmented_label = []            # y
print('Begin!!')

#TODO: CHANGE!
for person in ECG:        # read 'lead' data (i.e. MLIII)
    k = 0
    whole_signal=wfdb.rdrecord(rootdir + '/' + person).p_signal.transpose()     # whole signal
    while (k+1)*f*segmented_len<=len(whole_signal[0]):    # until last data point, index with segments
        count+=1
        # read a record without .xxx
        record = wfdb.rdrecord(rootdir + '/' + person, sampfrom=k * f * segmented_len,sampto=(k + 1) * f * segmented_len)  
        # read .atr for annotation
        annotation = wfdb.rdann(rootdir + '/' + person, 'atr', sampfrom=k * f * segmented_len,sampto=(k + 1) * f * segmented_len)  
        #TODO: CHANGE!
        lead_index = record.sig_name.index('ECG')  # find the index
        signal = record.p_signal.transpose()  # transpose for figure
        #print(signal.shape)
        label = []           # record all corresponding labels, and pick from here
        #signal_rounded = np.round(signal, 3)    # round to 3 decimal
        #segmented_data.append(signal_rounded[lead_index])   # if sample rate is 360, no need to resample
        symbols = annotation.symbol
 
        re_signal = scipy.signal.resample(signal[lead_index], 3600)  # resampling to 360 x 10 secs = 3600
        re_signal_rounded = np.round(re_signal, 3)    # round to 3 decimal
        print('resignal', re_signal_rounded)
        segmented_data.append(re_signal_rounded)

        # segmented_data.append(re_signal)
        print('symbols', symbols, len(symbols))


        if '+' in symbols:  # delete +
            symbols.remove('+')
        if len(symbols) == 0:   # None label
            segmented_label.append('Q')
        # if all 'N'or'/'and'N' combinationmark as N
        elif symbols.count('N') / len(symbols) == 1 or symbols.count('N') + symbols.count('/') == len(symbols):  
            segmented_label.append('N')
        else:
            for i in symbols:
                if i != 'N':
                    label.append(i)
            segmented_label.append(label[0])

        # print(label)
        k+=1
print('Begin to save dataset!')


# segmented_data=pd.DataFrame(segmented_data)
# segmented_label=pd.DataFrame(segmented_label)
# #TODO: CHANGE!
# segmented_data.to_csv('X_MBST_ECG.csv', index=False)
# segmented_label.to_csv('Y_MBST_ECG.csv', index=False)

print('Finished!')

```
